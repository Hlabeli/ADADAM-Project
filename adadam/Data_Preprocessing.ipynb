```python
import pandas as pd
import numpy as np
```

```python
col_names = ["duration", "protocol_type", "service", "flag", "src_byes",
            "dst_bytes", "land", "wrong_fragment", "urgent", "hot", "num_failed_logins",
            "logged_in", "num_compromised", "root_shell", "su_attempted", "num_root", "num_file_creations",
            "num_shells", "num_access_files", "num_bound_cmds", "is_host_login", "is_guest_login", "count",
            "srv_count", "serror_rate", "srv_serror_rate", "rerror_rate", "srv_rerror_rate", "same_srv_rate", "diff_srv_rate",
            "srv_diff_host_rate", "dst_host_count", "dst_host_srv_count", "dst_host_same_srv_rate", "dst_host_diff_srv_rate",
            "dst_host_same_src_port_rate", "dst_host_srv_diff_host_rate", "dst_host_serror_rate", "dst_host_srv_serror_rate",
            "dst_host_rerrror_rate", "dst_host_srv_rerror_rate", "label", "difficulty_level"]
```

```python
data = pd.read_csv('dataset/KDDTrain+.txt', header=None, names=col_names)
```

```python
data
```

```python
data.drop(['difficulty_level'], axis=1, inplace=True)
data.shape
```

```python
data.describe()
```

```python
data['label'].value_counts()
```

```python
def change_label(df):
    df.label.replace(['apache2', 'back', 'land', 'neptune', 'mailbomb', 'pod', 'processtable', 'smurf', 'teardrop', 'udpstorm',
                    'worm'], 'Dos', inplace=True)
    df.label.replace(['ftp_write', 'guess_passwd', 'httptunnel', 'imap', 'multihop', 'named', 'phf', 'sendmail', 
                      'snmpgetattack', 'snmpguess', 'spy', 'warezclient', 'warezmaster', 'xlock', 'xsnoop'], 'R2L', inplace=True)
    df.label.replace(['ipsweep', 'mscan', 'nmap', 'portsweep', 'saint', 'satan'], 'Probe', inplace=True)
    df.label.replace(['buffer_overflow', 'loadmodule', 'perl', 'ps', 'rootkit', 'sqlattack', 'xterm'], 'U2R', inplace=True)
```

```python
change_label(data)
```

```python
data.label.value_counts()
```

```python
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
```

```python
numeric_col = data.select_dtypes(include='number').columns
```

```python
std_scaler = StandardScaler()
def normalization(df,col):
    for i in col:
        arr = df[i]
        arr = np.array(arr)
        df[i] = std_scaler.fit_transform(arr.resharp(len(arr),1))
    return df
```

```python
data.head()
```

```python
data = normalization(data.copy(),numeric_col)
```

```python
cat_col = ['protocol_type', 'service', 'flag']
```

```python
categorical = data[cat_col]
categorical.head()
```

```python
categorical = pd.get_dummies(categorical, columns=cat_col)
categorical.head()
```

```python
bin_label = pd.DataFrame(data.label.map(lambda x:'normal' if x=='normal' else 'abnormal'))
```

```python
bin_data = data.copy()
bin_data['label'] = bin_label
```

```python
le1 = preprocessing.LabelEncoder()
enc_label = bin_label.apply(le1.fit_transform)
bin_data['intrusion'] = enc_label
```

```python
le1.classes_
```

```python
np.save('le1_classes.npy', le1.classes_, allow_pickle=True)
```

```python
bin_data.head()
```

```python
from pandas import get_dummies


bin_data=get_dummies(bin_data, columns=['label'], prefix="", prefix_sep="")
bin_data['label'] = bin_label
bin_data
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,8))
plt.pie(bin_data.label.value_counts(), labels=bin_data.label.unique(), autopct='%0.2f%%')
plt.title("Distrabution of normal and abnormal")
plt.legend()
plt.savefig('adadam/static/images/binary_pie_chart.png')
plt.show()
```
